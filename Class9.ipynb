## Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
plt.plot(x, y)
plt.title('A single plot')
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,4))
axes[1].plot(x, -y)
axes[0].plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x, y)
plt.title("figura de la izquierda")
plt.subplot(1,2,2)
plt.plot(x, -y)
plt.title("figura de la derecha")
plt.show()
```

```python
fig, (axes1, axes2) = plt.subplots(2,1,figsize=(5,8))
fig.suptitle('Vertically stacked subplots')
axes1.plot(x, y)
axes2.plot(x, -y)
plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.subplot(2,1,1)
plt.plot(x, y)
plt.title('Vertically stacked subplots')
plt.subplot(2,1,2)
plt.plot(x, -y)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2,figsize=(10,10))

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0,0]')
axs[0, 1].plot(x, y, 'orange')
axs[0, 1].set_title('Axis [0,1]')
axs[1, 0].plot(x, -y, 'green')
axs[1, 0].set_title('Axis [1,0]')
axs[1, 1].plot(x, -y, 'red')
axs[1, 1].set_title('Axis [1,1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(x, y)
plt.title('Axis [0,0]')
plt.subplot(2,2,2)
plt.plot(x, y, 'orange')
plt.title('Axis [0,1]')
plt.subplot(2,2,3)
plt.plot(x, -y, 'green')
plt.title('Axis [1,0]')
plt.subplot(2,2,4)
plt.plot(x, -y, 'red')
plt.title('Axis [1,1]')
plt.show()
```

## Several figures

```python
import matplotlib.pyplot as plt
plt.figure(1)      # the first figure
plt.subplot(211)   # the first subplot in the first figure
plt.plot([1,2,3])
plt.subplot(212)   # the second subplot in the first figure
plt.plot([4,20,55])

plt.figure(2)      # a second figure
plt.plot([-4,-10,-15])  # creates a subplot(111) by default


plt.figure(1)      # figure 1 current; subplot(212) still current
plt.subplot(211)   # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3')# subplot 211 title
plt.subplot(212)   # make subplot(211) in figure1 current
plt.plot([4,20,55],"r-x")
```

## Histograms

#### num_bins= 50

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

mu = 100    # mean of distribution
sigma = 15  # standard deviation of distribution
num_bins = 50
x = mu + sigma * np.random.randn(500)

n, bins, patches =plt.hist(x, num_bins, density=1)

# actual gaussian curve
gaussian = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((bins - mu)/ sigma )**2))

plt.plot(bins, gaussian, '--r')
plt.xlabel('Bins')
plt.ylabel('Probability density')
plt.title(r'Histogram: $\mu=100$, $\sigma=15$')
plt.show()
```

## Bar charts

#### Vertical Bar

```python
import numpy as np
import matplotlib.pyplot as plt

means_men = (20, 35, 30, 35, 27)
means_women = (25, 32, 34, 20, 25)
index = np.arange(len(means_men)) 
 
bar_width = 0.35
opacity = 0.4

plt.bar(index, means_men, bar_width,
                alpha=opacity, color='b',
                label='Men')

plt.bar(index + bar_width, means_women, bar_width,
                alpha=opacity, color='g',
                label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width / 2, labels=('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.show()
```

#### Horizontal bar

```python
import numpy as np
import matplotlib.pyplot as plt

means_men = (20, 35, 30, 35, 27)
means_women = (25, 32, 34, 20, 25)
index = np.arange(len(means_men)) 
 
bar_width = 0.35
opacity = 0.4

plt.barh(index, means_men, bar_width,
                alpha=opacity, color='b',
                label='Men')

plt.barh(index + bar_width, means_women, bar_width,
                alpha=opacity, color='g',
                label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.yticks(index + bar_width / 2, labels=('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.show()
```

## Timeseries (Financial data)

#### Remember to install datareader if required

If needed go to Anaconda console and run: "conda install pandas-datareader"

#### Get Financial data using package pandas-datareader

```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd

# Define the instruments to download. We would like to see Apple and Microsoft
tickers = ['AAPL', 'MSFT']

# We would like all available data from 01/01/2000 until 12/31/2016.
start_date = '2010-01-01'
end_date =   '2019-01-27'

# User pandas_reader.data.DataReader to load the desired data. As simple as that.
panel_data = data.DataReader(tickers, 'yahoo', start_date, end_date)
df=panel_data.stack()
df.head()
```

#### Focus on "Close" stock value, ignore the weekends and fill blank

```python
# Getting just the adjusted closing prices.
close = panel_data['Close']
print("Data received:", close.shape)
close.head(8)
```

#### Plot Microsfot Close graph

```python
# Get the MSFT timeseries. This now returns a Pandas Series object indexed by date.
msft = close.iloc[:]['MSFT']

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_msft = msft.rolling(window=20).mean()
long_rolling_msft  = msft.rolling(window=100).mean()

# Plot everything
plt.figure(figsize=(16,9))

plt.plot(msft.index, msft, label='MSFT')
plt.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling',color="red")
plt.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling',color="magenta")

plt.xlabel('Date')
plt.ylabel('Adjusted closing price ($)')
plt.legend()
plt.show()
```
